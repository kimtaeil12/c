1)데이터 준비, 불러오기, 확인

```python
# 라이브러리 불러오기

import pandas as pd
import tensorflow as tf
```

```python
# 데이터 불러오기

파일경로 = "/content/lemonade (1).csv"
레모네이드 = pd.read_csv("/content/lemonade (1).csv")
```

```python
# 변수안에 어떤 데이터가 있는지 확인 (데이터가 생겼는지 확인)

print(레모네이드.shape)
```

```python
# 컬럼 이름 출력

print(레모네이드.columns)
```

```python
# 독립 변수는 원인이고, 종속변수는 결과값이다

독립 = 레모네이드[['온도']]
종속 = 레모네이드[['판매량']]

print(독립.shape, 종속.shape)
```

```python
# 실제 데이터 출력

레모네이드.head
```

2)모델의 구조를 만듭니다

```python
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

3)모델을 학습시킵니다

```python
model.fit(독립, 종속, epochs=10000)
```

4)모델이 학습한 결과를 봅니다

```python
print('Predictions : ' , model.predict([[15]]))
```
